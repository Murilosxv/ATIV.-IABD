```python
from sklearn.model_selection import train_test_split # importa a biblioteca
# para dividir os dados em treino e teste
import numpy as np # biblioteca para trabalhar com arrays
from sklearn.tree import DecisionTreeClassifier # importa o modelo 
from sklearn import tree
#arvore de decisão
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns # importa biblioteca seaborn para plotar gráficos
```

```python
data = pd.DataFrame({
    'velocidade':[10,5,8,6,12,4],
    'Complexidade':[0,2,1, 2, 0, 2],
    'Manutencao': ['0','2','1','2','0','1'],
    'Classificacao':['Montagem','Teste','Montagem','Teste','Montagem','Teste']

    # 0 = baixa
    #  1 = média 
    #  2 = alta 

})
```

```python
x=np.array(data['velocidade']) # pega a coluna de tempo do dataframe
y=np.array(data['Classificacao']) # pega a coluna de classificação do dataframe 
x_treino,x_teste,y_treino,y_teste = train_test_split(x.reshape(-1,1),y.reshape(-1,1),
                                                     test_size=0.2,random_state=42)
model = DecisionTreeClassifier()
model.fit(x_treino,y_treino)
prev = model.predict(x_teste)
acurracy = accuracy_score(y_teste,prev)
precisao = precision_score(y_teste,prev,pos_label='Teste') # pós label rótulo
matrix = confusion_matrix(y_teste,prev)
print(f'Acurácia: {acurracy}')
print(f'Precisão: {precisao}')
fig = plt.figure(figsize=(10,8))
tree.plot_tree(model,feature_names=x.tolist(),class_names=y.tolist(),filled=True)
plt.savefig('EXERCÍCIO 01 (ÁrvoreDecisão).jpg')
plt.show()
```

```python
sns.heatmap(matrix,annot= True,fmt='d',cmap='Blues',cbar=False,xticklabels=['Teste','Montagem'],yticklabels=['Teste','Montage',])
plt.xlabel('Previsão')
plt.ylabel('Verdadeiro')
plt.title('Matriz de Confusão - Classificação das máquinas')
plt.savefig('EXERCÍCIO 01 (Matriz de Confusão).jpg')
plt.show()
```

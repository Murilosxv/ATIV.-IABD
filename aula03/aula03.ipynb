```python
import pandas as pd #Importa a biblioteca
import numpy as np 
import matplotlib.pyplot as plt
```

```python
dados_X = np.linspace(0,10,10)
dados_Y = dados_X * dados_X
```

```python
plt.figure(figsize=(10,8),facecolor='gray') # cria uma figura
Ax = plt.axes() # cria os eixos para o grafico
Ax.set_facecolor('White') # cor do fundo do grafico 
plt.plot(dados_X,dados_Y,
         color='red', marker= 'o',
         linestyle= 'dashed',
         markeredgecolor='black'
) # plota o grafico em x e y 
plt.xlabel("Eixo x",fontsize=16)
plt.ylabel("Eixo y",fontsize=16)
plt.title("Grafico simples com matplotlib",fontsize=20)
```

```python
plt.rcParams.update({'font.size':15})
plt.figure(figsize=(10,10))
linhas =2
colunas = 2
plt.subplot(linhas,colunas,1)
plt.plot(dados_X,'r--',label='Dados_x')
plt.legend(loc = 'lower right')
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Grfico 1")
plt.subplot(linhas,colunas,2)
plt.plot(dados_X,dados_Y,'b-.',label='Dados X \n e Y')
plt.legend(loc = 'lower right')
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Grfico 2")
plt.subplot(linhas,colunas,3)
plt.plot(dados_X*dados_Y,'g-',label='Dados X * Y')
plt.legend(loc = 'lower right')
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Grfico 3")
plt.subplot(linhas,colunas,4)
plt.plot(dados_Y,'c--',label='Dados Y')
plt.legend(loc = 'lower right')
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Grfico 4")
plt.show()
```

```python
#grafico de barras 
Figura = plt.figure()
grafico = Figura.add_axes([0,0,0.7,0.7]) # tamanhho do grafico
rotulos = [ 'A','B','C','D','E','F','G','H','I','J']
#Gerando gradico 
grafico.bar(rotulos, dados_X)
grafico.bar(rotulos, dados_X, color = 'red')
```

```python
grafico.bar(rotulos, dados_X, color = 'red')
grafico.bar(rotulos, dados_Y*0.1, color = 'blue')
plt.show
```

```python
import pandas as pd # importa a biblioteca pandas
```

```python
Dicionario_01={
    'ID':[1,2,3,4],
    'Nome':["Ana","Pedro","Maria","Murilo"],
    'Setores':["Rh","Vendas","Administrativo","Produção"]
}
Dicionario_02={
    'ID':[1,2,3,4],
    'Salarios':[3000,2800,3200,2500]
    }
```

```python
df1 = pd.DataFrame(Dicionario_01)
df2 = pd.DataFrame(Dicionario_02)
```

```python
df_n1 = pd.merge(df1,df2,how='inner', on ='ID')
```
